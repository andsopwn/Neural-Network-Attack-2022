```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import struct
def binary(num):
    return ''.join('{:0>8b}'.format(c) for c in struct.pack('!f', num))

dir = '../Trace/' # WSL 22.04.
#dir = '../../mnt/c/Users/louxsoen/Documents/P/Nerual-Network-SCA/Float-Point-NN-Mantissa/Trace/'
```

```python
cutsize = 200
relu = np.load(dir + 'relumany.npy')
input = np.linspace(-10, 10, cutsize)
#print(input)

hw  = np.zeros(200)
cut = np.zeros(200)

for i in range(1, 200):
    hw[i] = binary(input[i]).count("1")

plt.figure(figsize=(5,1))
plt.plot(hw)
print(input.shape)
print(relu.shape)
```

```python
corr = np.zeros(2000)
loc = 0
max = 0
for i in range(1, 2000):
    for j in range(1, 200):  cut[j] = relu[j][i]
    corr[i] = np.corrcoef(cut, hw)[0,1]
    if corr[i] > max:
        max = corr[i]
        loc = i

print(corr)
plt.plot(abs(corr))

print(max, loc)
```
